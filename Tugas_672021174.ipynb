Soal 3, Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

hid_penutup = ('Donat', 'Pastel', 'Kue Coklat', 'Ice Cream', 'Puding Vanila', 'Brownies', 'Puding Strawberry', 'Puding Coklat', 'Ice Cream Nutela', 'Kue Coklat-Keju', 'Kue Wajik', 'Kue Sus', 'Mochi')
data_penjualan = (14, 5, 12, 19, 6, 8, 12, 9, 10, 17, 2, 9, 13)

df = pd.DataFrame({
    'Dessert':hid_penutup,
    'Popular':data_penjualan,
})
df.sort_values(by='Popular', inplace=True)

x_coords = np.arange(len(df))
warna = ['#0000FF' for _ in range(len(df))]
warna[-11] = '#FF0000'
warna[-12] = '#FF0000'
warna[-13] = '#FF0000'
plt.figure(figsize=(10,5))
plt.bar(x_coords, df['Popular'], tick_label=df['Dessert'], color=warna)
plt.xticks(rotation = 90)
plt.ylabel('Terjual')
plt.title('Data Penjualan Hidangan Penutup')
plt.show()
```

Sebutkan tiga makanan penutup yang harus disingkirkan = Kue Wajik, Pastel, Puding Vanila
